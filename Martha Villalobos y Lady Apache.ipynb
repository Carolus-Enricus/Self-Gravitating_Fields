### 1 - Introducción y Solución para el Estado Inicial 
\begin{eqnarray}
 \partial_t \phi_1 &=& \frac{\alpha}{a}\pi_1 \\
 \partial_t \phi_2 &=& \frac{\alpha}{a}\pi_2 \\
 \partial_t \psi_1 &=& \partial_r \left( \frac{\alpha\pi_1}{a} \right)\\ \tag{Field Equations}
 \partial_t \psi_2 &=& \partial_r \left( \frac{\alpha\pi_2}{a} \right)\\
 \partial_t \pi_1 &=& \frac{1}{r^2}\partial_r\left(\frac{r^2\alpha\psi_1}{a}\right)-\alpha a\phi_1 \\
 \partial_t \pi_2 &=& \frac{1}{r^2}\partial_r\left(\frac{r^2\alpha\psi_2}{a}\right)-\alpha a\phi_2
\end{eqnarray}

\begin{eqnarray}
 \frac{a'}{a} &=& \frac{1-a^2}{2r} + \frac{r}{2}\left[ \psi_1^2+\psi_2^2 + \pi_1^2+\pi_2^2 + a^2(\phi_1^2+\phi_2^2) \right] \\ \tag{Einstein Field Equations}
 \frac{\alpha'}{\alpha} &=& \frac{a^2-1}{2r} + \frac{r}{2}\left[  \psi_1^2+\psi_2^2 + \pi_1^2+\pi_2^2 - a^2(\phi_1^2+\phi_2^2)  \right] \\
 \partial_t a &=& r\alpha\left(\psi_1\pi_1+\psi_2\pi_2\right)\\
\end{eqnarray}

Relación entre las variables:

$$\pi_i = \dfrac{a}{\alpha}\partial_t \phi_i \qquad \psi_i = \partial_r \phi_i$$

El truco para evitar problemas con la singularidad en $r=0$:

$$ \partial_t \pi_i = 3\frac{\partial}{\partial r^3}\left(\frac{r^2\alpha\psi_i}{a}\right)-\alpha a\phi_i $$

Se piensa en el campo para tiempos cortos como un campo armónico en el tiempo. De modo que la condición inicial se propone de la siguiente manera:

$$ \phi(r)e^{-i\omega t} $$

Yisus dijo que era como pensar en ángulos pequeños para un péndulo. Ésto nos da a las nuevas variables como:

\begin{eqnarray}
 \phi_1 &=& \phi(r)\cos(\omega t) \qquad \qquad \phi_2 = -\phi(r)\sin(\omega t) \\
 \pi_1 &=& -\omega \dfrac{a}{\alpha}\phi(r)\sin(\omega t) \qquad \pi_2 = -\omega \dfrac{a}{\alpha}\phi(r)\cos(\omega t)\\
 \psi_1 &=& \phi'(r)\cos(\omega t) \qquad \qquad \psi_2 = -\phi'(r)\sin(\omega t)
\end{eqnarray}

Utilizando Mathematica se encuentra que:

\begin{eqnarray}
 \phi' &=& \psi \\
 \psi' &=& -\left(\dfrac{2}{r} + \dfrac{\alpha'}{\alpha} - \dfrac{a'}{a}  \right)\psi + a^2\left(1 - \dfrac{\omega^2}{\alpha^2}  \right)\phi \\
 \dfrac{a'}{a} &=& \dfrac{1-a^2}{2r} + \dfrac{r}{2}\left[ \psi^2 + a^2\left(1+\dfrac{\omega^2}{\alpha^2} \right)\phi^2 \right] \\
   \dfrac{\alpha'}{\alpha} &=& \dfrac{a^2-1}{2r} + \dfrac{a'}{a} -ra^2\phi^2
\end{eqnarray}

Hay un truco bastante sucio que no se menciona en los artículos: que para las ecuaciones de $a$ y de $\alpha$ se evaluó explícitamente en $t=0$. 

Las condiciones de frontera son:
\begin{eqnarray}
 \phi(0) &=& \phi_0 \qquad  \phi(\infty) &=& 0\\ 
 \psi(0) &=& 0 \qquad \psi(\infty) &=& 0\\
 a(0) &=& 1 \qquad a(\infty) &=& 1 \\
 \alpha(0) &=& \alpha_0 \qquad \alpha(\infty) &=& \dfrac{1}{a(\infty)} \\ 
\end{eqnarray}

Las condiciones para las funciones métricas son tales que en infinito se obtenga la métrica de Schwarzschild.

Luego, para arrancar el Runge-Kutta es necesario determinar tres constantes: $\omega$, $\phi_0$ y $\alpha_0$. Con las siguientes definiciones:

$$\alpha(r) = \alpha_0 \hat{\alpha}(r)$$
$$\omega = \alpha_0 \hat{\omega} $$

Tenemos el siguiente sistema de ecuaciones

\begin{eqnarray}
 \phi' &=& \psi \\
 \psi' &=& -\left(\dfrac{2}{r} + \dfrac{\hat{\alpha}'}{\hat{\alpha}} - \dfrac{a'}{a}  \right)\psi + a^2\left(1 - \dfrac{\hat{\omega}^2}{\hat{\alpha}^2}  \right)\phi \\
 \dfrac{a'}{a} &=& \dfrac{1-a^2}{2r} + \dfrac{r}{2}\left[ \psi^2 + a^2\left(1+\dfrac{\hat{\omega}^2}{\hat{\alpha}^2} \right)\phi^2 \right] \\
   \dfrac{\hat{\alpha}'}{\hat{\alpha}} &=& \dfrac{a^2-1}{2r} + \dfrac{a'}{a} -ra^2\phi^2
\end{eqnarray}

Con sus respectivas condiciones de frontera:

\begin{eqnarray}
 \phi(0) &=& \phi_0 \qquad  \phi(\infty) &=& 0\\ 
 \psi(0) &=& 0 \qquad \psi(\infty) &=& 0\\
 a(0) &=& 1 \qquad a(\infty) &=& 1 \\
 \hat{\alpha}(0) &=& 1 \qquad \hat{\alpha}(\infty) &=& \dfrac{1}{a(\infty)\alpha_0} \\ 
\end{eqnarray}

Y en este sistema sólo hay que determinar dos parámetros de arranque. Luego, si queremos hallar la solución de $\alpha$ basta con reconstruirla tomando:

$$\alpha_0 = \dfrac{1}{\hat{\alpha}[end]*a[end]} \qquad \alpha = \alpha_0 \hat{\alpha}$$

### 2 - Masa ADM y $\sigma$

La relación entre de estas funciones con la masa $m(r)$ y la función $\sigma(r)$ utilizada por Daka, se obtiene de comparar las siguientes definiciones:

$$ ds^2 = -\alpha^2(r)dt^2 + a^2dr^2 +r^2 d\Omega^2 $$
$$ ds^2 = -N(r)\sigma(r)^2(r)dt^2 + \dfrac{dr^2}{N(r)} +r^2 d\Omega^2 $$

$$ N(r) := 1 - \dfrac{2m(r)}{r} $$
$$ \sigma(r) := \alpha(r)a(r) $$

$$ m(r) = \dfrac{r}{2}\left( 1 - \dfrac{1}{a(r)^2} \right) $$

````julia
#--------------------------------------------------#
#                   Ejecución #1                   #
#--------------------------------------------------#

using Plots
using Roots
using SpecialFunctions

# Sistema de Ecuaciones
function deq_System(w, A, r)
    a, alpha, phi, psi = A[1], A[2], A[3], A[4]
    
    dln_a = (1-a^2)/(2*r) + 0.5*r*(psi^2 + (a*phi)^2*(1+(w/alpha)^2)) # Derivada logarítmica de a
    dln_alpha = (a^2-1)/r + dln_a - r*(a*phi)^2
    dphi = psi
    dpsi = -(2/r + dln_alpha - dln_a)*psi + a^2*(1-(w/alpha)^2)*phi

    return [a*dln_a, alpha*dln_alpha, dphi, dpsi]
end

# Parámetros del dominio

Nr = 1200 # Número de pasos
rf = 30 # Extremo derecho del dominio
h = rf / Nr # Tamaño del pasito perrón

#--------------------------------#
#---Solucionador Runge-Kutta-4---#
#--------------------------------#
function Static_Solver(f0,w)

    lr = Float64[] # Malla  
    la = Float64[] # Aquí albergamos la solución  de a
    lalpha_gorro = Float64[] # Aquí albergamos la solución  de alpha
    lphi = Float64[] # Aquí albergamos a tu papá alcohólico, que diga! a la solución  de psi :D
    lpsi = Float64[] # Aquí albergamos la solución  de phi  
    
    # Condiciones de Frontera:
    r = h
    A = [1.0, 1.0, f0, 0.0]

        for i in 1:Nr

            push!(lr,r)
            push!(la, A[1])  
            push!(lalpha_gorro, A[2])
            push!(lphi, A[3])
            push!(lpsi, A[4])

            r += h

            k1=h*deq_System(w, A, r)
            k2=h*deq_System(w, A+0.5*k1, r+0.5*h)
            k3=h*deq_System(w, A+0.5*k2, r+0.5*h)
            k4=h*deq_System(w, A+k3, r+h)

            A += (k1+2*k2+2*k3+k4)/6
        end
    
    return [lr, la, lalpha_gorro, lphi, lpsi]
end

#----------------------#
#---Función de error---#
#----------------------#

function error(f0,w)
    error = Static_Solver(f0,w)[4][end]
    return error
end    

#-----------------#
#---Root finder---#
#-----------------#

# Esta función de julia encontrará los ceros de la función error. Aquí empieza la acción.
f0 = 0.2
w_seed = 1.1639438950879792 # Esta es la semilla que permite arrancar el buscador de raíces

w_guess = find_zero((w) -> error(f0,w), w_seed)
println(w_guess) # Esta es la omega que logra calcular con precisión de maquina

# Valor al que tiende alpha gorro en infinito
alpha0 = 1/(Static_Solver(f0, w_guess)[3][end]*Static_Solver(f0, w_guess)[2][end])

# Omega "original"
w = alpha0*w_guess
println(w)


#---Campos y Métrica------------------

r = Static_Solver(f0, w_guess)[1]
a = Static_Solver(f0, w_guess)[2]
alpha = [alpha0*k for k in Static_Solver(f0, w_guess)[3]] # Reconstrucción de alpha
phi = Static_Solver(f0, w_guess)[4]
psi = Static_Solver(f0, w_guess)[5]

#---Masa y Sigma----------------------

m = [i/2*(1-1/j^2) for (i, j) in zip(r, a)] # Masa
sigma = [i*j for (i, j) in zip(a, alpha)] # Sigma 

#---------------------------------------------------------------------------------------
#                                           |
#                                           |
#                 .''.         .''. `._    _|_    _.' .''.         .''.
#                  '. '.     .' .'     ~-./ _ \.-~     '. '.     .' .'
#  ____              '. '._.' .'         /_/_\_\         '. '._.' .'               ____
# '.__ ~~~~-----......-'.' '.'`~-.____.-~       ~-..____.-~'.' '.'`-......-----~~~~ __.'
#     ~~~~----....__//.''._.'.\\\       GRAFIACION      ///.'._.''.\\___....----~~~~
#                   .' .'__'. '._..__               __.._.' .'__'. '.
#                 .' .'||    '. '.   ~-.._______..-~   .' .'    ||'. '.
#                '.,'  ||-.    ',.'        |_|        '.,'    .-||  ',.'
#                      \| |                .'.                | |/
#                       | |                | |                | |
#                       '.'                '.'                '.'

plt_static = plot(layout = (1, 3), size = (900, 300))


plot!(plt_static[1], r, phi, xlabel="r", label = raw"\phi (r)")
plot!(plt_static[1], r, psi, xlabel="r", label = raw"\psi (r)", linecolor=:green)

plot!(plt_static[2], r, a, xlabel="r", label = "a(r)", linecolor=:purple)
plot!(plt_static[2], r, alpha, xlabel="r", label = "α(r)", linecolor=:red)

plot!(plt_static[3], r, m, xlabel="r", label=raw"m(r)", linecolor=:brown)
plot!(plt_static[3], r, sigma, xlabel="r", label=raw"\sigma (r)", linecolor=:green)

display(plt_static)
````

### 5 - Discretización de las Ecuaciones

Tenemos las condiciones iniciales para el tiempo $t=0$ para todo $r$:

$$ \phi^0_{1j}, \quad \psi^0_{1j}, \quad \pi^0_{1j} $$ 
$$ \phi^0_{2j}, \quad \psi^0_{2j}, \quad \pi^0_{2j} $$
$$ a^0_j, \quad \alpha^0_j $$

La discretización de las ecuaciones se da de la siguiente manera:

\begin{equation}\tag{1}
\phi^{n+1}_{1j} = \phi^{n}_{1j} +\Delta t \left[ \dfrac{\alpha^{n}_{j}}{a^{n}_{j}} \pi^{n}_{1j} \right]
\end{equation}

\begin{equation}\tag{2}
\phi^{n+1}_{2j} = \phi^{n}_{2j} +\Delta t \left[ \dfrac{\alpha^{n}_{j}}{a^{n}_{j}} \pi^{n}_{2j} \right]
\end{equation}

\begin{equation}\tag{3}
\psi^{n+1}_{1j} = \psi^{n}_{1j} +\Delta t \dfrac{1}{2\Delta r}\left[ \dfrac{\alpha^{n}_{j+1}}{a^{n}_{j+1}}\pi^{n}_{1j+1} - \dfrac{\alpha^{n}_{j-1}}{a^{n}_{j-1}}\pi^{n}_{1j-1} \right]
\end{equation}

\begin{equation}\tag{4}
\psi^{n+1}_{2j} = \psi^{n}_{2j} +\Delta t \dfrac{1}{2\Delta r}\left[ \dfrac{\alpha^{n}_{j+1}}{a^{n}_{j+1}}\pi^{n}_{2j+1} - \dfrac{\alpha^{n}_{j-1}}{a^{n}_{j-1}}\pi^{n}_{2j-1} \right]
\end{equation}

Para $\pi$ aplicamos el truco de derivar respecto a $r^3$ para los primeros pasos espaciales

\begin{equation}\tag{5}
\pi^{n+1}_{1j} = \pi^{n}_{1j} +\Delta t \left[\dfrac{3}{(r_{j+1})^3 - (r_{j-1})^3 }\left( \dfrac{\alpha^{n}_{j+1}}{a^{n}_{j+1}} \left(r_{j+1}\right)^2\psi^{n}_{1j+1} - \dfrac{\alpha^{n}_{j-1}}{a^{n}_{j-1}} \left(r_{j-1}\right)^2\psi^{n}_{1j-1} \right) -a^{n}_j\alpha^{n}_j\phi^{n}_{1j}\right]
\end{equation}

\begin{equation}\tag{6}
\pi^{n+1}_{2j} = \pi^{n}_{2j} +\Delta t \left[\dfrac{3}{(r_{j+1})^3 - (r_{j-1})^3 }\left( \dfrac{\alpha^{n}_{j+1}}{a^{n}_{j+1}} \left(r_{j+1}\right)^2\psi^{n}_{2j+1} - \dfrac{\alpha^{n}_{j-1}}{a^{n}_{j-1}} \left(r_{j-1}\right)^2\psi^{n}_{2j-1} \right) -a^{n}_j\alpha^{n}_j\phi^{n}_{2j}\right]
\end{equation}

Y ya luego derivamos normal:

\begin{equation}\tag{7}
\pi^{n+1}_{1j} = \pi^{n}_{1j} +\Delta t * \left[ \dfrac{1}{ 2\Delta r (r^2_j) }\left( \dfrac{\alpha^{n}_{j+1}}{a^{n}_{j+1}} \left(r_{j+1}\right)^2\psi^{n}_{1j+1} - \dfrac{\alpha^{n}_{j-1}}{a^{n}_{j-1}} \left(r_{j-1}\right)^2\psi^{n}_{1j-1} \right)-a^{n}_j\alpha^{n}_j\phi^{n}_{1j} \right]
\end{equation}

\begin{equation}\tag{8}
\pi^{n+1}_{2j} = \pi^{n}_{2j} +\Delta t * \left[ \dfrac{1}{ 2\Delta r (r^2_j) }\left( \dfrac{\alpha^{n}_{j+1}}{a^{n}_{j+1}} \left(r_{j+1}\right)^2\psi^{n}_{2j+1} - \dfrac{\alpha^{n}_{j-1}}{a^{n}_{j-1}} \left(r_{j-1}\right)^2\psi^{n}_{2j-1} \right)-a^{n}_j\alpha^{n}_j\phi^{n}_{2j} \right]
\end{equation}

Si acaso $r^n_{j-1} = 0$ o muy cercano a $0$ no habría problema porque entra el truco de derivar respecto a $r^3$ y el denominador no explota. Para la métrica tenemos:

\begin{equation}\tag{9}
a^n_{j+1}=a^n_{j}+\Delta r * a^n_j\left[ \dfrac{1-(a^n_j)^2}{2r_j}+\dfrac{r_j}{2} \left[ (\psi^n_{1j})^2+ (\psi^n_{2j})^2 + (\pi^n_{1j})^2 + (\pi^n_{2j})^2 + (a^n_j)^2((\phi^n_{1j})^2+(\phi^n_{2j})^2) \right] \right]
\end{equation}

\begin{equation}\tag{10}
\alpha^n_{j-1}=\alpha^n_{j}-\Delta r * \alpha^n_j\left[ \dfrac{(a^n_j)^2-1}{2r_j}+\dfrac{r_j}{2} \left[ (\psi^n_{1j})^2+ (\psi^n_{2j})^2 + (\pi^n_{1j})^2 + (\pi^n_{2j})^2 - (a^n_j)^2((\phi^n_{1j})^2+(\phi^n_{2j})^2) \right] \right]
\end{equation}

Estas dos últimas ecuaciones pueden resolverse con Runge-Kutta. En cada paso temporal se resuelven.

\begin{equation}\tag{7}
adot^{n+1}_{j}=\dfrac{a^{n+1}_j-a^n_j}{\Delta t}-r_j\alpha^n_j\left(\psi^n_{1j}\pi^n_{1j} + \psi^n_{2j}\pi^n_{2j} \right)
\end{equation}

````julia
#------------------------------------------#       ⠀⠀⠀⠀⠀⠀⢀⣠⣤⣴⣶⣶⠿⠿⠿⠿⠿⠿⢶⣶⣦⣤⣄⡀⠀⠀⠀⠀⠀⠀
#                Ejecución #4              #          ⢀⣴⣾⠿⠛⠉⠁⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠉⠛⠿⣷⣦⡀
#------------------------------------------#       ⢀⣴⡿⠋⠀⠀⠀            ⠀⠀⠀⠀⠀⠀⠀ ⠙⢿⣦⡀⠀
#                                                ⢠⣿⠋⠀⠀⠀                       ⠀ ⠀⠙⣿⡄
#⠀⠀                                             ⣾⡏⠀⠀⠀ Ésta es una herramienta     ⢸⣷
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀                       ⣿⡇⠀   que⠀utilizaremos más tarde ⠀⢸⣿
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢠⣴⣾⣿⣿⣿⣷⣦⡀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠸⣿⡄⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢠⣿⠇
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣰⣿⣿⣿⣿⣿⣿⣿⣿⣿⡆⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠙⢿⣦⡀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢀⣴⡿⠋⠀⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡇⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠙⣿⡆⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣀⣤⣶⠿⠋⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢿⣿⣿⣿⣿⣿⣿⣿⣿⡟⠁⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ ⢰⣿⠀⠀⠀⠀⠀⢀⣶⣦⣤⣤⣤⣤⣴⣶⣶⠿⠿⠛⠉⠀⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠙⠿⢿⣿⣿⣿⣿⣿⣶⣶⠶⠦⢤⣄⡀⠀⠀⠀⠀⠀⠀⠀⠀   ⣠⣿⠃⠀⢀⣠⣤⣾⠟⠋⠈
#⠀⠀⠀⠀⠀⠀⠀⠀⢀⣀⣄⡀⠀⠀⠀⠀⣠⣾⣿⣿⣿⣿⡿⠁⠀⠀⠀⠙⢿⢦⡀⠀⠀⠀⠀⠀⠀⠀⠀⢿⣷⡾⠿⠟⠛⠉⠀⠀
#⠀⠀⠀⠀⠀⠀⣠⣾⣿⣿⣿⣿⣷⣦⢀⣾⣿⣿⣿⣿⣿⣿⡃⠀⠀⠀⠀⢠⠈⠀⢳⡄⠀⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⣰⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡇⠀⠀⠀⠀⣾⣇⠀⣷⢳⠀⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⠟⠛⠛⠛⠻⣷⠀⠀⠀⠀⢻⣿⠀⣿⣼⣧⣠⣾⣷⡄
#⠀⠀⠀⠀⠀⠹⣿⣿⣿⣿⣿⣿⣿⣿⣿⡟⠀⠀⢀⣤⡤⠤⠈⠻⠀⠀⠀⠸⠛⠒⠉⠀⢠⣿⣿⡿⠃
#⠀⠀⠀⠀⠀⠀⠙⢿⣿⣿⣿⣿⠟⠁⠸⣧⠀⠀⠀⢸⣿⣄⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠈⠻⣿⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠉⠁⠀⠀⠀⠀⠈⠷⣀⢰⣾⣿⣿⣷⣄⡀⠀⠀⠀⠀⠀⠀⠀⣀⠾⠁⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠈⣹⣧⢻⣿⣿⣿⠿⣳⣶⣶⣶⡒⠒⠋⠁⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢀⣴⣿⡿⠛⠳⣶⣾⣿⣿⣿⣿⣿⣿⣷⣤⣀⡀⠀⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢠⣾⣿⡿⠃⠀⠀⢹⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡆⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣴⣿⣿⣿⡁⠀⠀⣀⣼⣿⣿⣿⣿⣿⣿⣿⣿⣿⠟⢻⣿⡄⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠘⠿⣿⣿⣿⣷⣾⣿⣿⣿⣿⣿⣿⣿⠟⠻⣿⣿⠀⢸⣿⡇⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠉⢻⣿⣿⣿⣿⣿⣿⣿⣿⣿⠀⠀⣿⣿⣧⣾⣿⠁⠀     (Utilizaré estas funciones para resolver las
#⠒⠒⠒⠶⠦⢄⣀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢸⣿⣿⣿⣿⣿⣿⣿⣿⣿⣦⣴⣿⣿⣿⣿⠃⠀⠀     constricciones métricas dentro de la evolución)
#⠀⠀⠀⠀⠀⠀⠈⠙⠲⣤⡀⠀⠀⠀⠀⠀⠀⠀⠀⣹⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡆⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠉⠲⢤⣀⠀⠀⠀⠀⢾⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡿⠟⠉⠀⠀⠀
#⠀⠀⠀⠀⠀⠀⠀⣠⡤⠴⠲⢦⣄⡈⠙⠒⠒⠒⠋⢙⣿⣿⣿⣿⣿⣿⡟⢹⣿⣿⡿⠀⠀⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⢀⣾⠋⣀⠀⠀⠀⠈⠻⡄⠀⠀⠀⣠⣾⣿⣿⠃⠀⠉⠉⢡⣾⣿⡿⠁⠀⠀⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⢸⡟⠉⠉⠛⢦⡀⠀⠀⠿⡆⢀⣴⣿⣿⡿⠁⠀⠀⠀⣰⣿⣿⡟⠁⠀⠀⠀⠀⠀⠀⠀
#⠀⠀⠀⠀⠀⠀⣿⡆⠀⠀⠸⡻⣆⠀⠀⢻⣿⣿⣿⠟⠀⢀⡴⢻⣿⣿⡿⣿⣷⠴⠚⠛⠛⠛⠶⣄⠀
#⠀⠀⠀⠀⠀⠀⠘⢧⣰⠀⠀⠐⠹⣦⠀⠀⠛⣬⡏⠀⠀⣹⠷⣬⠀⠠⠾⠋⠀⠀⠀⠀⠀⠀⠀⡿⡆
#⠀⠀⠀⠀⠀⠀⠀⠈⢻⣄⣀⣀⣀⣈⣷⣄⠀⠸⡇⠀⢸⡇⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣤⣿⠃
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠉⠻⣯⣉⠉⠁⠈⢳⣦⡇⠀⠀⠙⠶⣤⣀⣀⣀⣀⣀⣀⣀⣤⣤⠾⠟⠁⠀
#⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠉⠛⠛⠛⠛⠋⠀
# Ecuación para a
function deq_a_evolve(phi1, phi2, psi1, psi2, pi1, pi2, a, r) # Sigue sin depender de alpha :) 
    dln_a = (1-a^2)/(2*r) + 0.5*r*(psi1^2 + psi2^2 + pi1^2 + pi2^2 + a^2*(phi1^2+phi2^2))
    return a * dln_a
end

# Ecuación para alpha
function deq_alpha_evolve(phi1, phi2, psi1, psi2, pi1, pi2, alpha, a, r) # Sí depende de alpha
    dln_alpha = (a^2-1)/(2*r) + 0.5*r*(psi1^2 + psi2^2 + pi1^2 + pi2^2 - a^2*(phi1^2+phi2^2))
    return alpha * dln_alpha
end
````

````julia
#--------------------------------------------------#
#                   Ejecución #5                   #
#--------------------------------------------------#
using DelimitedFiles
#--------------------------------#
#---Archivos de Almacenamiento---#
#--------------------------------#
datos_phi1 = []
datos_phi2 = []

datos_a = []
datos_alpha = []
datos_adot = []

# El siguiente archivo se utilizará para estudiar el valor central del campo en el tiempo
datos_CentralValuePhi1 = []

# Aquí mero inicia el Ciclo Principal de Evolución
t = 0.0

# Crear la figura inicial fuera del ciclo
plt_evolve = plot(xlabel="r", size=(500, 400))

for i in 1:Nt # Los Geht's!
    
    #------------------------#
    #---Escritura de Datos---#
    #------------------------#
    if mod(i, 60) == 0 # Escribiremos cada 30 datos
        for j in 1:Nr
            push!(datos_phi1, [t, r[j], phi1[j]])
            push!(datos_phi2, [t, r[j], phi2[j]])
            push!(datos_a, [t, r[j], a[j]])
            push!(datos_alpha, [t, r[j], alpha[j]])
            push!(datos_adot, [t, r[j], adot[j]])
        end
    end
    
    # Almacenamos el tiempo y el valor de phi(0) a ese tiempo.
    push!(datos_CentralValuePhi1, [t, phi1[1]])
    
    #----------------------#
    #---Fuentes de Ortiz---#
    #----------------------#
    
    #---Para phi1----------------------------
    for j in 1:Nr
        s_phi1[j] = alpha[j]*pi1[j]/a[j]
    end
    
    #---Para phi2----------------------------
    for j in 1:Nr
        s_phi2[j] = alpha[j]*pi2[j]/a[j]
    end
    
    #---Para psi1----------------------------
    for j in 2:Nr-1
        s_psi1[j] = 0.5/dr * (alpha[j+1]*pi1[j+1]/a[j+1] - alpha[j-1]*pi1[j-1]/a[j-1])
    end
    
    #---Para psi2----------------------------
    for j in 2:Nr-1
        s_psi2[j] = 0.5/dr * (alpha[j+1]*pi2[j+1]/a[j+1] - alpha[j-1]*pi2[j-1]/a[j-1])
    end
        
    #---Truco de derivar respecto a r³ pi1-----
    for j in 2:11
        s_pi1[j] = 3.0/(r[j+1]^3 - r[j-1]^3)*(alpha[j+1]*r[j+1]^2*psi1[j+1]/a[j+1] - alpha[j-1]*r[j-1]^2*psi1[j-1]/a[j-1]) - a[j]*alpha[j]*phi1[j]
    end
    
    #---Truco de derivar respecto a r³ pi2-----
    for j in 2:11
        s_pi2[j] = 3.0/(r[j+1]^3 - r[j-1]^3)*(alpha[j+1]*r[j+1]^2*psi2[j+1]/a[j+1] - alpha[j-1]*r[j-1]^2*psi2[j-1]/a[j-1]) - a[j]*alpha[j]*phi2[j]
    end
    
    #---Resto de la RHS para pi1----------------
    for j in 12:Nr-1
        s_pi1[j] = 0.5/dr *(alpha[j+1]*r[j+1]^2*psi1[j+1]/a[j+1] - alpha[j-1]*r[j-1]^2*psi1[j-1]/a[j-1])/r[j]^2 - a[j]*alpha[j]*phi1[j]  
    end
    
    #---Resto de la RHS para pi2----------------
    for j in 12:Nr-1
        s_pi2[j] = 0.5/dr *(alpha[j+1]*r[j+1]^2*psi2[j+1]/a[j+1] - alpha[j-1]*r[j-1]^2*psi2[j-1]/a[j-1])/r[j]^2 - a[j]*alpha[j]*phi2[j]  
    end
        
    #-----------------------------#                           #        .'''.
    #---Truco del Pasito Perrón---#                           #       '     '
    #-----------------------------#                           #       '.   .'
                                                              #         '.'
    phi1_p = phi1                                             #          ' 
    psi1_p = psi1                                             #          '
    pi1_p = pi1                                               #          '
                                                              #        /")\
    phi2_p = phi2                                             #       / /\ \
    psi2_p = psi2                                             #      / ()""()    
    pi2_p = pi2                                               #     |=( . .)
                                                              #     |  '(o)|
    a_p = a                                                   #     /______|
                                                              #     |      |
    #-----------------------------#                           #   _ / ____ \ _
    #---Evolución de los campos---#                           #  ( ) /    \ ( )
    #-----------------------------#                           #  |_|/      \|_|
    
    # En este condicional se programa el ICN

    if i<50                                                     #        __      _
        dtw = 0.5*dt                                            #        \.'---.//|
    else                                                        #         |\./|  \/
        dtw = dt                                                #        _|.|.|_  \
    end                                                         #       /(  ) ' '  \
                                                                #      |  \/   . |  \
    #---- Evolución de las phi's-------------                   #      \_/\__/| |
    for j in 1:Nr                                               #        V  /V / |
        phi1[j] = phi1_p[j]+dtw*s_phi1[j]                       #          /__/ /
        phi2[j] = phi2_p[j]+dtw*s_phi2[j]                       #          \___/\
    end
    
    #---- Evolución de las psi's y pi's-------------
    
    for j in 2:Nr-1
        psi1[j] = psi1_p[j] + dtw*s_psi1[j]
        psi2[j] = psi2_p[j] + dtw*s_psi2[j]
        pi1[j] = pi1_p[j] + dtw*s_pi1[j]
        pi2[j] = pi2_p[j] + dtw*s_pi2[j]
    end
    
    #-----------------------------#
    #---Condiciones de Frontera---#
    #-----------------------------#
    
    psi1[1] = -psi1[2]
    psi2[1] = -psi2[2]
    pi1[1] = pi1[2]
    pi2[1] = pi2[2]
    
    # Condición del Compact
    dtfac = dtw/dr
    aux = dtw/(r[Nr] + r[Nr-1])
    
    pi1[Nr] = (pi1[Nr-1]*(dtfac - 1.0 - aux) + pi1_p[Nr]*(1.0 - dtfac - aux) + pi1_p[Nr-1]*(1.0 + dtfac - aux))/(1.0 + dtfac + aux)
    pi2[Nr] = (pi2[Nr-1]*(dtfac - 1.0 - aux) + pi2_p[Nr]*(1.0 - dtfac - aux) + pi2_p[Nr-1]*(1.0 + dtfac - aux))/(1.0 + dtfac + aux)
    
    psi1[Nr] = - pi1[Nr] - phi1[Nr]/rf
    psi2[Nr] = - pi2[Nr] - phi2[Nr]/rf
    
    #-----------------------------#
    #---Constricciones Métricas---#
    #-----------------------------#
    a[1]=1.0
  
    for j in 1:Nr-1 # Este ciclo soluciona a
        k1 = dr * deq_a_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], a[j], r[j])
        k2 = dr * deq_a_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], a[j] + 0.5*k1, r[j] + 0.5*dr)
        k3 = dr * deq_a_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], a[j] + 0.5*k2, r[j] + 0.5*dr)
        k4 = dr * deq_a_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], a[j] + k3, r[j] + dr)
        
        a[j+1] = a[j] + (k1 + 2*k2 + 2*k3 + k4)/6
    end

    alpha[Nr] = 1/a[Nr]

    for j in reverse(2:Nr)
        k1 = dr * deq_alpha_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], alpha[j], a[j], r[j])
        k2 = dr * deq_alpha_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], alpha[j]+ 0.5*k1, a[j], r[j] + 0.5*dr)
        k3 = dr * deq_alpha_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], alpha[j]+ 0.5*k2, a[j], r[j] + 0.5*dr)
        k4 = dr * deq_alpha_evolve(phi1[j], phi2[j], psi1[j], psi2[j], pi1[j], pi2[j], alpha[j]+ k3, a[j], r[j] + dr)
        
        alpha[j-1] = alpha[j] - (k1 + 2*k2 + 2*k3 + k4)/6
    end
    
    #---- La molesta adot -------------
    for j in 1:Nr-1
        adot[j] = (a[j]-a_p[j])/dt - 0.5*r[j]*alpha[j]*(psi1[j]*pi1[j] + psi2[j]*pi2[j])
    end
    
    #---- a² y alpha²-------------
    
    a_sqrt = [k^2 for k in a]
    alpha_sqrt = [k^2 for k in alpha]
    
    #----------------------------#
    #---Visualización de Datos---#
    #----------------------------#
    
    impresiones = 100
    
    if mod(i, impresiones) == 0
        plot!(plt_evolve, r, a_sqrt, label = false)
        plot!(plt_evolve, r, alpha_sqrt, label = false)
    end
    
    t += dt # Y ya sólo actualizamos el tiempo
end

display(plt_evolve)

writedlm("phi1.dat", datos_phi1)
writedlm("phi2.dat", datos_phi2)
writedlm("a.dat", datos_a)
writedlm("alpha.dat", datos_alpha)
writedlm("adot.dat", datos_adot)
writedlm("phi1_0.dat", datos_CentralValuePhi1)
````

### 6 - Transformada de Fourier del valor central del campo

Tomaremos el corte a $r=0$ a modo de obtener una función $\phi_1(0,t)$ y obtener su transformada de fourier.

````julia
#--------------------------------------------------#                           #              .''
#                   Ejecución #6                   #                           #    ._.-.___.' (`\
#--------------------------------------------------#                           #   //(        ( `'
using FFTW                                                                     #  '/ )\ ).__. ) 
# Esta librería contiene a la Transformada Fourier más rápida del Oeste        #  ' <' `\ ._/'\
                                                                               #     `   \     \
# Procedemos a llamar y leer el archivo que contiene el valor central de phi en cada paso de tiempo
archivo = "phi1_0.dat"
datos = readdlm(archivo)
tiempo = datos[:, 1]
phi1_0 = datos[:, 2]

# En esta parte calculamos la transformada
transformada = fft(phi1_0)
frecuencias = fftfreq(length(tiempo), 1 / (tiempo[2] - tiempo[1]))

# Aquí tomamos de la transformada las amplitudes y las partes real e imaginaria
amplitudes = abs.(transformada) # Amplitudes
parte_real = real.(transformada) # Parte real
parte_imaginaria = imag.(transformada) # Parte Imaginaria

#-----------------#
#---Graficación---#
#-----------------#
plt_fourier = plot(layout = (2,2), size = (900, 600))

plot!(plt_fourier[1], tiempo, phi1_0, xlabel = "t", label = raw"\phi_1 (0)")
plot!(plt_fourier[2], frecuencias, amplitudes, xlabel = "w", label = "Amplitudes")
scatter!(plt_fourier[2], [1 / (2 * π)], [0], label = "Frecuencia Característica", color = :black, marker = :utriangle) 
plot!(plt_fourier[3], frecuencias, parte_real, xlabel = "w", label = "Real Part")
plot!(plt_fourier[4], frecuencias, parte_imaginaria, xlabel = "w", label = "Imaginary Part")

display(plt_fourier)
````
