### 1 - Introduction and Solution for the Initial State
\begin{eqnarray}
 \partial_t F_1 &=& -\dfrac{\alpha}{r}\left(G_1-rF_2 \right) -\sqrt{\dfrac{\alpha}{a}} \partial_r\left(G_1\sqrt{\dfrac{\alpha}{a}}\right) \\
 \partial_t F_2 &=& -\dfrac{\alpha}{r}\left(G_2+rF_1 \right) -\sqrt{\dfrac{\alpha}{a}} \partial_r\left(G_2\sqrt{\dfrac{\alpha}{a}}\right)  \\
 \partial_t G_1 &=& \dfrac{\alpha}{r}\left(F_1-rG_2 \right) -\sqrt{\dfrac{\alpha}{a}} \partial_r\left(F_1\sqrt{\dfrac{\alpha}{a}}\right)  \\ \tag{Field Equations}
 \partial_t G_2 &=& \dfrac{\alpha}{r}\left(F_2+rG_1 \right) -\sqrt{\dfrac{\alpha}{a}} \partial_r\left(F_2\sqrt{\dfrac{\alpha}{a}}\right) 
\end{eqnarray}

\begin{eqnarray}\tag{Einstein Field Equations}
 \frac{\alpha'}{\alpha} &=& \frac{a^2-1}{2r} + r a^2 S_r^r \\ 
  \frac{a'}{a} &=& \frac{1-a^2}{2r} + r a^2 \rho \\ 
\end{eqnarray}

Where $\rho$ and $S_r^r$ are releated with the energy-momentum tensor and have a physical interpretation:

$$ S_r^r = \dfrac{2}{r^2a^2} \left[ (F_1G'_2-F_2G'_1) + (G_1F'_2-G_2F'_1) \right] $$

$$ \rho = \dfrac{2}{r^2a^2} \left[ a(F_1^2+F_2^2-G_1^2-G_2^2) + \dfrac{2a}{r}(F_1G_2-F_2G_1) + (F_1G'_2-F_2G'_1) + (G_1F'_2-G_2F'_1) \right] $$

Static solution means a solution where space-time is time independent. For this to be possible, the components of the energy momentum tensor must be time independent and $T_{tr}$ must vanish. For a field of the next form:

$$ F(t,r) = f(r) e^{-i\omega t} \qquad G(t,r) = ig(r) e^{-i\omega t}$$

this can be achieved. With these assumtions, we get:

\begin{eqnarray}
F_1 = f(r)\cos(\omega t) \qquad F_2 = -f(r)\sin(\omega t)\\
G_1 = g(r)\sin(\omega t) \qquad G_2 = g(r)\cos(\omega t)
\end{eqnarray}

And the partial differential equations system yield to an ordinary differential equations system. In Daka, the equations are written in terms of variables such as $m$, $N$, $S_r^r$, etc. In order to work in terms of $a$ and $\alpha$, just as it was done for scalar fields, Im going to modify the Daka's equations and use the next quantities:

$$ r=\dfrac{\bar{r}}{\mu} \qquad t=\dfrac{\bar{t}}{\mu} \qquad \omega=\mu \bar{\omega} \qquad m=\dfrac{1}{\mu G}\bar{m} $$

$$ f = \dfrac{1}{\sqrt{\mu G}}\bar{f} \qquad g = \dfrac{1}{\sqrt{\mu G}}\bar{g} $$

$$ \rho = \dfrac{\mu^2}{4\pi G}\bar{\rho} \qquad S_r^r = \dfrac{\mu^2}{4\pi G}\bar{S}_r^r$$

God willing, I got the correct static equations:

\begin{eqnarray}
 f' &=& -a\left( \dfrac{\omega}{\alpha}+1 \right)g +f \left[ \dfrac{a}{r} - \dfrac{1}{2} \left(\dfrac{\alpha'}{\alpha} - \dfrac{a'}{a}\right)\right] \\
 g' &=& a\left( \dfrac{\omega}{\alpha}-1 \right)f - g \left[ \dfrac{a}{r} + \dfrac{1}{2} \left(\dfrac{\alpha'}{\alpha} - \dfrac{a'}{a}\right)\right] \\
\end{eqnarray}

\begin{eqnarray}
S_r^r = \dfrac{2}{r^2a^2} \left( fg' - f'g \right) \\
\rho = \dfrac{2\omega}{a\alpha r^2}(f^2+g^2) 
\end{eqnarray}



Up to this point, the equations for each variable, are in terms of derivatives of $f$ and $g$ and that is a problem when implementing the Runge-Kutta method. This problem is solved in an easy way using the equations for $f$ and $g$ and finding that:

$$ fg'-f'g = a\left[ \dfrac{\omega}{\alpha}(f^2+g^2) +g^2-f^2 \right] - \dfrac{2a}{r}fg $$

So, $\rho$ and $S^r_2$ can be determinated in terms of non-derivative functions, for that reason the equations for $a$ and $\alpha$ have the same property. To apply the Runge-Kutta method, we need boundary conditions. For the metric, we use the fact that in the infinity we recover the space-time of Schwarzschild. I took the conditions for $f$ and $g$ in Herdeiro, who can argue that in $r=0$, $g=0$.

\begin{eqnarray}
 f(0) &=& f_0 \qquad f(\infty) = 0\\ 
 g(0) &=& 0 \qquad  g(\infty) = 0\\
 a(0) &=& 1 \qquad a(\infty) = 1 \\
 \alpha(0) &=& \alpha_0 \qquad \alpha(\infty) = \dfrac{1}{a(\infty)} \\ 
\end{eqnarray}

As it was just made for K-G Field, we'll define two new variables in order to have only one constant for finding:

$$ \alpha = \alpha_0 \hat{\alpha} \qquad \omega = \alpha_0 \hat{\omega} $$

We'll have essentially the same system equation but with these conditions:

\begin{eqnarray}
 f(0) &=& f_0 \qquad f(\infty) = 0\\ 
 g(0) &=& 0 \qquad  g(\infty) = 0\\
 a(0) &=& 1 \qquad a(\infty) = 1 \\
 \alpha(0) &=& 1 \qquad \alpha(\infty) = \dfrac{1}{\alpha_0 a(\infty)} \\ 
\end{eqnarray}

Finally, we recover the solution for $\alpha$ computing this:

$$\alpha_0 = \dfrac{1}{\hat{\alpha}[end]*a[end]} \qquad \alpha = \alpha_0 \hat{\alpha}$$

````julia
#--------------------------------------------------#
#                   Ejecution #1                   #
#--------------------------------------------------#

using Plots
using Roots
using SpecialFunctions

# System Equation
function deq_System(w, A, r)
    a, alpha, f, g = A[1], A[2], A[3], A[4]
    
    caborca = a*((w/alpha)*(f^2+g^2) + g^2 - f^2) - (2*a*f*g)/r # This variable is fg'-f'g but I wanted call it like this  
    
    Srr = (2/(r*a)^2) * caborca # SHEAR STRESS
    rho = (2*w/(a*alpha*r^2)) * (f^2+g^2) # DENSITY
    
    dln_alpha = (a^2-1)/(2*r) + r*a^2*Srr # Logarithmic derivative of the alpha
    dln_a = (1-a^2)/(2*r) + r*a^2*rho # Logarithmic derivative of a
    
    df = -a*g*(w/alpha + 1) + f*(a/r - 0.5(dln_alpha - dln_a))
    dg =  a*f*(w/alpha - 1) - g*(a/r + 0.5(dln_alpha - dln_a))

    return [a*dln_a, alpha*dln_alpha, df, dg]
end

# Parameters of the space domain

Nr = 1500 # Number of points in the grid
rf = 15 # Right bounday of the domain
h = rf / Nr # Size of the step

#--------------------------#
#---Runge-Kutta-4 method---#
#--------------------------#
function Static_Solver(f0,w)

    lr = Float64[] # Grid
    la = Float64[] # Here solution for a is saved
    lalpha_gorro = Float64[] # Here solution for alpha is saved
    lf = Float64[] # Here we save your alcoholic father, Sorry! the solution of f :D
    lg = Float64[] # And this array if for g
    
    # Boundary conditions:
    r = h
    A = [1.0, 1.0, f0, 0.0]

        for i in 1:Nr

            push!(lr,r)
            push!(la, A[1])  
            push!(lalpha_gorro, A[2])
            push!(lf, A[3])
            push!(lg, A[4])

            r += h

            k1=h*deq_System(w, A, r)
            k2=h*deq_System(w, A+0.5*k1, r+0.5*h)
            k3=h*deq_System(w, A+0.5*k2, r+0.5*h)
            k4=h*deq_System(w, A+k3, r+h)

            A += (k1+2*k2+2*k3+k4)/6
        end
    
    return [lr, la, lalpha_gorro, lf, lg]
end

#--------------------#
#---Error Function---#
#--------------------#

function error(f0,w)
    error = Static_Solver(f0,w)[4][end]
    return error
end    

#-----------------#
#---Root finder---#
#-----------------#

# This function of julia is going to find the zeros of the error function. 

f0 = 0.01
w_seed = 1.8757330495966809 # This is the god seed 

w_guess = find_zero((w) -> error(f0,w), w_seed)
println(w_guess) # This is the omega with computer precision

# Valor al que tiende alpha gorro en infinito
alpha0 = 1/(Static_Solver(f0, w_guess)[3][end]*Static_Solver(f0, w_guess)[2][end])

# "Original" omega before the existance of alpha0
w = alpha0*w_guess
println(w)

#---Campos y Mtrica------------------

r = Static_Solver(f0, w_guess)[1]
a = Static_Solver(f0, w_guess)[2]
alpha = [alpha0*k for k in Static_Solver(f0, w_guess)[3]] # Reconstruccin de alpha
f = Static_Solver(f0, w_guess)[4]
g = Static_Solver(f0, w_guess)[5]

#---Masa y Sigma----------------------

m = [i/2*(1-1/j^2) for (i, j) in zip(r, a)] # Masa
sigma = [i*j for (i, j) in zip(a, alpha)] # Sigma 

#--------------------------------------------------------------------
#       _.--,_......----..__
#       \  .'          '    ```--...__      \
#        \;           '            .  '.   ||
#        :           '            '     \ .''.
#      .';          :            '       .|  |.--..___
#     /   \         |  PLOTING  :        :|  /.------.\
#    /    .'._      :           |        ||  ||      |\\
#   /.-. /|-| `-.               :        ;|  ||______|_\`.______
#  //  // |-|    \   '           '      / |  ||='      | |      `.
# //  //\\|-|     `-._'           '   .'  |  ||        | |        \
#/.-.//  \\-|_________```------------` ___'. ||        | '_.--.   <)
#'._.'  /  .-----.   .-----.   .''''''''.    |'--..____| /  _  \   |
#       |_/.'   '.\_/.'   '.\_[ [ [  ] ] ]___|_________.'.'   '.\  ]
#         :  .-.  : :  .-.  :  '........'    (_________):  .-.  :`-'
#         :  '-'  : :  '-'  :                           :  '-'  :
#          '._ _.'   '._ _.'                             '._ _.'

plt_static = plot(layout = (1, 3), size = (900, 300))


plot!(plt_static[1], r, f, xlabel="r", label = raw"f (r)")
plot!(plt_static[1], r, g, xlabel="r", label = raw"g (r)", linecolor=:green)

plot!(plt_static[2], r, a, xlabel="r", label = "a(r)", linecolor=:purple)
plot!(plt_static[2], r, alpha, xlabel="r", label = "(r)", linecolor=:red)

plot!(plt_static[3], r, m, xlabel="r", label=raw"m(r)", linecolor=:brown)
plot!(plt_static[3], r, sigma, xlabel="r", label=raw"\sigma (r)", linecolor=:green)

display(plt_static)
````
